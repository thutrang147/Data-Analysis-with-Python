```python
class TextAnalyzer:
    def __init__(self, file_name):
        with open(f"text_file/{file_name}", "r", encoding="utf-8") as f:
            self.text = f.read()
    
    def clean_text(self, to_lower=True):
        self.normalize_text = self.text
        if to_lower:
            self.normalize_text = self.text.lower()
        self.normalize_text = re.sub(r'[^a-zA-Z0-9\s\n]', ' ', self.normalize_text)

        words = self.normalize_text.split()
        return words
    
    def word_frequency(self, words, ascending=False):
        self.words_freq = {}
        for word in words:
            if word not in self.words_freq:
                self.words_freq[word] = 1
            else:
                self.words_freq[word] += 1
        if ascending == True:
            sorted_words_freq = dict(sorted(self.words_freq.items(), key=lambda item: item[1]))
        else: 
            sorted_words_freq = dict(sorted(self.words_freq.items(), key=lambda item: item[1], reverse= not ascending))
            
        return sorted_words_freq
```

```python
class ExtendedTextAnalyzer(TextAnalyzer):
    def __init__(self, file_name):
        super().__init__(file_name)

    def unique_words(self):
        frequency = self.word_frequency(self.clean_text())
        uniquewords = sorted(frequency.keys())
        return uniquewords
    
    def sentence_vectors(self):
        dictionary = self.word_frequency(self.clean_text())
        dictionary = dict(list(dictionary.items()))
        
        self.map_index = {word: idx for idx, word in enumerate(dictionary.keys())}
        self.words = set(self.map_index.keys())
        self.vector_size = len(dictionary)

        sentences = self.text.split('.')
        text_matrix = []
        for sentence in sentences:
            vector = [0] * self.vector_size
            words = self.clean_text(sentence)
            for word in words:
                if word in self.words:
                    vector[self.map_index[word]] += 1
            if sum(vector) != 0:
                text_matrix.append(vector)
            
        return text_matrix, sentences
    
    def most_similar_sentences(self, sentencevectors):
        norm_vectors = np.array(sentencevectors)
        norms = np.linalg.norm(norm_vectors, axis=1, keepdims=True)
        norm_vectors = norm_vectors / (norms + 1e-9)

        dot_product = np.dot(norm_vectors, norm_vectors.T)

        max_sim = 0
        max_i, max_j = -1, -1
        for i in range(1, len(dot_product)):
            for j in range(i + 1, len(dot_product)):
                if dot_product[i][j] > max_sim:
                    max_sim = dot_product[i][j]
                    max_i, max_j = i, j

        return max_i, max_j
    
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for d in range(3, int(n**0.5) + 1, 2):
        if n % d == 0:
            return False
    return True
```

```python
def is_even(n):
    if n % 2 != 0:
        return False
    return True
```

```python
def entries_larger_than_mean(matrix):
    mean_value = np.mean(matrix)
    count = np.sum(matrix > mean_value)
    return count

def most_freq(matrix):
    unique, counts = np.unique(matrix, return_counts=True)
    max_count = np.max(counts)
    most_frequent = unique[counts == max_count]
    return most_frequent, max_count

def prime_number(matrix):
    is_prime_vectorized = np.vectorize(is_prime)
    prime_count = np.sum(is_prime_vectorized(matrix))
    return prime_count

def get_prime_factors(n):
    factors = []
    d = 2
    while n > 1:
        while n % d == 0:
            factors.append(d)
            n = n // d
        d += 1

    return factors

def is_product(n):
    factors = get_prime_factors(n)
    return len(factors) == 2 and all(is_prime(f) for f in factors)

def find(matrix):
    check1 = np.vectorize(is_even)
    check_vectorized = np.vectorize(is_product)
    mask1 = check1(matrix)
    mask = check_vectorized(mask1)

    if np.any(mask):
        return sum(matrix[mask]), matrix[mask], [get_prime_factors(n) for n in matrix[mask]]
    else:
        return f"There is no entry which is product of two prime numbers and is even number"

print("Count_entries", entries_larger_than_mean(M))
print("Most frequency", most_freq(M))
print("Number of prime", prime_number(M))
print("Sum of all elements:", find(M))
```

```python
def total_revenue(bill):
    total = 0
    for invoice in bill.split('\n'):
        items = re.findall(r'\((\d+\.\d+)\)', invoice)
        discount = re.search(r'Discount: (\d+)%', invoice)
        tax = re.search(r'Tax: (\d+)%', invoice)
        
        if items:
            subtotal = sum(float(item) for item in items)
            if discount:
                subtotal -= subtotal * (float(discount.group(1)) / 100)
            if tax:
                subtotal += subtotal * (float(tax.group(1)) / 100)
            total += subtotal
    
    return total

print(total_revenue(bill))
```

```python
def highest_total_invoice(bill):
    highest_total = 0
    highest_customer = ""
    for invoice in bill.split('\n'):
        items = re.findall(r'\((\d+\.\d+)\)', invoice)
        discount = re.search(r'Discount: (\d+)%', invoice)
        tax = re.search(r'Tax: (\d+)%', invoice)
        customer = re.search(r'Customer: ([\w\s]+)', invoice).group(1)
        
        if items:
            subtotal = sum(float(item) for item in items)
            if discount:
                subtotal -= subtotal * (float(discount.group(1)) / 100)
            if tax:
                subtotal += subtotal * (float(tax.group(1)) / 100)
            if subtotal > highest_total:
                highest_total = subtotal
                highest_customer = customer
    
    return highest_customer, highest_total

print(highest_total_invoice(bill))
```

```python
def product_sold_most(bill):
    products = re.findall(r'([\w\s]+) \(\d+\.\d+\)', bill)
    product_count = {}
    
    for product in products:
        if product not in product_count:
            product_count[product] = 1
        else:
            product_count[product] += 1
    
    if not product_count:
        return "No products found", 0
    
    most_sold_product = max(product_count, key=product_count.get)
    return most_sold_product, product_count[most_sold_product]

print(product_sold_most(bill))
```

```python
def day(bill):
    invoices = bill.split('\n')
    daily_totals = {}

    for invoice in invoices:
        date = re.search(r'Date: (\d{4}-\d{2}-\d{2})', invoice).group(1)
        day = date.split('-')[2]
        items = re.findall(r'\((\d+\.\d+)\)', invoice)
        discount = re.search(r'Discount: (\d+)%', invoice)
        tax = re.search(r'Tax: (\d+)%', invoice)

        if items:
            subtotal = sum(float(item) for item in items)
            if discount:
                subtotal -= subtotal * (float(discount.group(1)) / 100)
            if tax:
                subtotal += subtotal * (float(tax.group(1)) / 100)
            
            if day not in daily_totals:
                daily_totals[day] = subtotal
            else:
                daily_totals[day] += subtotal

    highest_day = max(daily_totals, key=daily_totals.get)
    return highest_day, daily_totals[highest_day]

print(day(bill))
```
