### Question 2. (40 points)

a) (20 points) Write a class called `DocumentProcessor`. This class should have the following methods:

- __init__(self, file): Initializes the object by reading the content from the specified file (located in the `Documents` folder) with UTF-8 encoding.
- clean_text(self, to_lower=True): Removes all non-alphanumeric characters (leaving only letters, numbers, and spaces) using a regular expression. If `to_lower` is set to `True` (default), converts the text to lowercase. Returns a list of cleaned words.
- word_histogram(self, word_list, descending=True): Takes a list of words and an optional parameter `descending` (default: `True`) to sort the word frequency histogram in descending order. Returns a dictionary where keys are words and values are their frequencies.
- Create a `DocumentProcessor` instance with `file = get_file(id)` (using the `get_file` function from `midterm_dseb65b.py`). Display the word frequency histogram by applying `clean_text` and `word_histogram`.

b) (20 points) Extend the `DocumentProcessor` class by creating a subclass called `AdvancedDocumentProcessor` with the following additional methods:
- unique_word_list(self): Returns an alphabetically sorted list of unique, cleaned words (based on `clean_text`).
- text_to_matrix(self): Represents each sentence in the file’s content as a row in a matrix, where each element of the row indicates the frequency of a word (from the list in `unique_word_list`) in that sentence.
- closest_sentence_pair(self): Calculates the Euclidean distance between all pairs of sentence vectors (from `text_to_matrix`) and returns the pair of sentences with the smallest Euclidean Distance score.
- Apply all methods of the `AdvancedDocumentProcessor` class using the `file` from part (a). Display each result accordingly.

```python
class DocumentProcessor:
    def __init__(self, file):
        with open(f'text_file/{file}', 'r', encoding='utf-8') as f:
            self.content = f.read()

    def clean_text(self, to_lower=True):
        text = re.sub(r'\n', '', self.content)
        text = re.sub(r'[^A-Za-z0-9\s]', '', self.content)
        if to_lower:
            text = text.lower()
        return text.split()

    def word_histogram(self, word_list, descending=True):
        histogram = {}
        for word in word_list:
            histogram[word] = histogram.get(word, 0) + 1
        return dict(sorted(histogram.items(), key=lambda x: x[1], reverse=descending))
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
import re

class AdvancedDocumentProcessor(DocumentProcessor):
    def unique_word_list(self):
        return sorted(set(self.clean_text()))

    def text_to_matrix(self):
        unique_words = self.unique_word_list()
        word_index = {word: i for i, word in enumerate(unique_words)}
        
        sentences = re.split(r'[.!?]+', self.content.strip())
        sentences = [sent.strip() for sent in sentences if sent.strip()]
        matrix = np.zeros((len(sentences), len(unique_words)), dtype=int)

        for i, sentence in enumerate(sentences):
            for word in sentence.lower().split():
                if word in word_index:
                    idx = word_index[word]
                    matrix[i, idx] += 1
        return matrix, sentences

    def closest_sentence_pair(self):
        matrix, sentences = self.text_to_matrix()
        distances = squareform(pdist(matrix, 'euclidean'))
        np.fill_diagonal(distances, np.inf) 
        min_idx = np.unravel_index(np.argmin(distances), distances.shape)
        sentence1, sentence2 = sentences[min_idx[0]], sentences[min_idx[1]]
        min_distance = distances[min_idx]
        return sentence1, sentence2, min_distance


adv_processor = AdvancedDocumentProcessor(file)

print("Unique words:", adv_processor.unique_word_list())
matrix, sentences = adv_processor.text_to_matrix()
print("Text matrix shape:", matrix.shape)

sentence1, sentence2, distance = adv_processor.closest_sentence_pair()
print("The closest sentence pair is:")
print("1:", sentence1)
print("2:", sentence2)
print("With a Euclidean distance of:", distance)
```

```python
class DocumentProcessor2:
    def __init__(self, file_name):
        with open(f"text_file/{file_name}") as file:
            self.text = file.read()
            
    def clean_text(self, text = None, lower=True):
        if text:
            self.normalize_text = text
        else:
            self.normalize_text = self.text
        if lower:
            self.normalize_text = self.normalize_text.lower()
        self.normalize_text = re.sub(r'[^a-zA-Z0-9\s]', ' ', self.normalize_text)

        words = self.normalize_text.split()
        return words
    
    def word_histogram (self, words, decreasing=True):
        frequency = {}
        for word in words:
            word = word.strip()
            if word in frequency:
                frequency[word] += 1
            else:
                frequency[word] = 1
        
        if decreasing:
            pairs =  sorted(frequency.items(), key=lambda x: x[1], reverse=True)
            frequency = dict(pairs)
            
        return frequency
```

```python
import numpy as np
import re

class AdvancedDocumentProcessor(DocumentProcessor2):
    def __init__(self, file_name):
        super().__init__(file_name)
        
    def unique_word_list(self):
        frequency = self.word_histogram(self.clean_text())
        unique_words = sorted(frequency.keys())
        return unique_words
    
    def text_to_matrix(self, vector_length=20000):
        dictionary = self.word_histogram(self.clean_text())
        dictionary = dict(list(dictionary.items())[:vector_length])
        
        self.map_index = {word: idx for idx, word in enumerate(dictionary.keys())}
        self.words = set(self.map_index.keys())
        self.vector_size = len(dictionary)
        
        sentences = self.text.split('.')
        text_matrix = []
        for sentence in sentences:
            vector = [0] * self.vector_size
            words = self.clean_text(sentence)
            for word in words:
                if word in self.words:
                    vector[self.map_index[word]] += 1
            if sum(vector) != 0:
                text_matrix.append(vector)
                
        return text_matrix, sentences

    def closest_sentence_pair(self, sentence_vectors):
        norm_vectors = np.array(sentence_vectors)
        norms = np.linalg.norm(norm_vectors, axis=1, keepdims=True)
        norm_vectors = norm_vectors / (norms + 1e-9)
        
        dot_product = np.dot(norm_vectors, norm_vectors.T)
        
        max_sim = 0
        max_i, max_j = -1, -1
        for i in range(1, len(dot_product)):
            for j in range(i + 1, len(dot_product)):
                if dot_product[i][j] > max_sim:
                    max_sim = dot_product[i][j]
                    max_i, max_j = i, j
                    
        return max_i, max_j
```

```python
B =    np.array([[ 9, 48, 30, 11, 84, 52, 94, 43, 53],
       [60, 83, 10, 89, 26, 85, 21, 53, 91],
       [ 9, 39, 86, 54, 73,  5, 44, 90, 67],
       [90, 59, 23, 97,  2, 17, 62, 75, 93],
       [71, 33, 65, 43, 97, 54, 72, 28, 98],
       [17, 89, 82, 17, 31,  0,  6, 95, 55],
       [90,  9, 83, 43, 21, 39,  6, 55, 99],
       [ 3, 71, 74, 50, 49,  4,  7, 42, 67]])
```

```python
from midterm_dseb65a import get_matrix
A = get_matrix(id)

def count_below_median(matrix):
    median = np.median(matrix)
    print(median)
    return np.sum(matrix < median)

def most_frequent_entry(matrix):
    unique, counts = np.unique(matrix, return_counts=True)
    max_count = np.max(counts)  
    most_frequent = unique[counts == max_count]  
    return most_frequent, max_count  

def count_perfect_squares(matrix):
    return np.sum([int(np.sqrt(x))**2 == x for x in matrix.flatten()])

def sum_distinct_prime_product(matrix):
    def is_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True

    def is_distinct_prime_product(x):
        primes = []
        for i in range(2, x):
            if x % i == 0 and is_prime(i):  
                other_factor = x // i
                if other_factor != i and is_prime(other_factor):  
                    primes.extend([i, other_factor])
                    return True
        return False

    return np.sum([x for x in matrix.flatten() if is_distinct_prime_product(x)])

print("Entries below median:", count_below_median(A))
print("Most frequent entry:", most_frequent_entry(A))
print("Perfect squares count:", count_perfect_squares(A))
print("Sum of distinct prime products:", sum_distinct_prime_product(A))
```

d) (5 points) Write a function called get_top_in_kfastest_above_threshold(…, k=10, score_threshold=5.0) to find the individual with the highest score among the 10 fastest individuals who have scores above a specified threshold (defaulting to 5.0 points).

```python
from midterm_dseb65b import get_test

test = get_test(id)
test
```

```python
def fix(data):
    return 0.67 if data == "missing" else 0.00 if data == "-" else float(data) if isinstance(data, str) else data


def calculate_total(test):
    return {row['ID']: sum(fix(row.get(f"Q{i}", 0)) for i in range(1, 16)) for row in test}


def parse_time_taken(time_str):
    parts = time_str.split()
    minutes = int(parts[0]) * 60  
    seconds = int(parts[2]) if len(parts) > 2 and 'sec' in parts[2] else 0
    return minutes + seconds

def get_name(student_id, id_dict):
    email = student_id if '@' in student_id else f"{student_id}@st.neu.edu.vn"
    return id_dict.get(email, email)  


def get_topk(test_data, id_dict, k=10):
    totals = calculate_total(test_data)
    sorted_totals = sorted(totals.items(), key=lambda x: x[1], reverse=True)
    topk = [
        (
            student_id,
            get_name(student_id, id_dict), 
            score
        )
        for student_id, score in sorted_totals[:k]
    ]
    return topk


def get_top_in_kfastest_above_threshold(test, k=10, score_threshold=5.0):
    scores = calculate_total(test)
    qualified = [(row['ID'], scores[row['ID']], parse_time_taken(row['Time taken'])) 
                 for row in test if scores[row['ID']] > score_threshold]
    top_fastest = sorted(qualified, key=lambda x: x[2])[:k]
    print(top_fastest)
    return max(top_fastest, key=lambda x: x[1]) if top_fastest else None

print("Top 10:", get_topk(test, id_dict))
print("Top in k fastest above threshold:", get_top_in_kfastest_above_threshold(test))
```
