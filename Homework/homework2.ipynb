```python
def minimum_path(triangle):
    n = len(triangle)
    if n == 0:
        return 0, []
    
    dp = [[0] * len(row) for row in triangle]
    dp[-1] = triangle[-1][:]

    for i in range(n - 2, -1, -1):
        for j in range(len(triangle[i])):
            dp[i][j] = triangle[i][j] + min(dp[i+1][j], dp[i+1][j+1])

    path = [triangle[0][0]]
    j = 0
    for i in range(1, n):
        if(dp[i][j] < dp[i][j+1]):
            path.append(triangle[i][j])
        else:
            path.append(triangle[i][j+1])
            j += 1
    
    return dp[0][0], path

import ast

triangle_in = input()
triangle = ast.literal_eval(triangle_in)

min_sum, path = minimum_path(triangle)
path_str = " -> ".join(map(str, path))
print(f"Minimum path: {path_str}, Sum = {min_sum}")

```
