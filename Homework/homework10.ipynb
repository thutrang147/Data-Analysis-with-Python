```python
# Student Grade Management
class StudentGrade():
    def __init__(self, course_name, grade):
        self.course_name = course_name
        self.grade = grade

    def average(self):
        return sum(self.grade) / len(self.grade)
    
    def highest(self):
        return max(self.grade)
    
    def lowest(self):
        return min(self.grade)
    
student = StudentGrade("Math", [10, 9, 8, 8])
print("Average Grade:", student.average())
print("Highest Grade:", student.highest())
print("Lowest Grade:", student.lowest())
```

```python
# Invoice Management
class Invoice():
    def __init__(self, invoice_number, created_date, total_amount):
        self.invoice_number = invoice_number
        self.created_date = created_date
        self.total_amount = total_amount

    def total(self):
        return sum(self.total_amount)
    
    def display(self):
        return f"The {self.invoice_number} created at {self.created_date} has the total of {self.total()}"
    
my_invoice = Invoice("123", "11/10/2024", [10, 20, 30])
print(my_invoice.display())
```

```python
# Address Book Management
class AddressBook:
    def __init__(self):
        self.full_name = []
        self.phone_number = []

    def add(self, name, phone):
        self.full_name.append(name)
        self.phone_number.append(phone)

    def edit_name(self, old_name, new_name):
        if old_name in self.full_name:
            index = self.full_name.index(old_name)
            self.full_name[index] = new_name
        else:
            print(f"Contact {old_name} not found.")
    
    def edit_phone(self, old_phone, new_phone):
        if old_phone in self.phone_number:
            index = self.phone_number.index(old_phone)
            self.phone_number[index] = new_phone
        else:
            print(f"Phone number {old_phone} not found")

    def delete(self, name):
        if name in self.full_name:
            index = self.full_name.index(name)
            self.full_name.pop(index)
            self.phone_number.pop(index)
        else:
            print(f"Contact {name} not found.")

    def display(self):
        for name, phone in zip(self.full_name, self.phone_number):
            print(f"Name: {name}, Phone: {phone}")

my_address = AddressBook()
my_address.add("Trang", "147")
my_address.add("Thu", "321")
my_address.add("Khanh", "054")
my_address.add("Qtr", "247")

my_address.edit_name("Khanh", "Linh")
my_address.edit_phone("321", "271")

my_address.delete("Qtr")

my_address.display()
```

```python
# Bank Account Management
class BankAccount():
    def __init__(self, account_number, account_holder, balance):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, money):
        if money > 0:
            self.balance += money
            return f"Deposited {money}. New balance is {self.balance}."
        else:
            return "Deposit amount must be positive."
    
    def withdraw(self, money):
        if money > 0:
            if self.balance >= money:
                self.balance -= money
                return f"Withdrew {money}. New balance is {self.balance}."
            else:
                return "Insufficient funds."
        else:
            return "Withdrawal amount must be positive."
    
    def check(self):
        return f"The current balance of the account number {self.account_number} is {self.balance}"

my_bank = BankAccount("1472005", "Trang", 100)
print(my_bank.deposit(200))
print(my_bank.check())
print(my_bank.withdraw(20))
print(my_bank.check())
```

```python
# Graphics Management
class Graphics():
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def peremeter(self):
        return (self.length + self.width) * 2
    def area(self):
        return self.length * self.width
    def draw(self):
        for _ in range(self.length):
            print('*' * self.width)
    
graphic = Graphics(5, 4)
print("Peremeter:", graphic.peremeter())
print("Area:", graphic.area())
graphic.draw()
```

```python
class Rectangle():
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __str__(self):
        return f"Rectangle({self.width}, {self.height})"
    
    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return self.width * self.height == other.width * other.height
        return False
    
    
    def __lt__(self, other):
        if isinstance(other, Rectangle):
            return self.width * self.height < other.width * other.height
        return False
            
    def __add__(self, other):
        if isinstance(other, Rectangle):
            new_width = self.width + other.width
            new_height = self.height + other.height
            return Rectangle(new_width, new_height)
        return NotImplemented
    

rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 10)
rect3 = Rectangle(3, 6)

print(rect1)  
print(rect1.__eq__(rect2))
print(rect1.__lt__(rect3)) 
print(rect1.__add__(rect2)) 
```
