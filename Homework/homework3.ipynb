```python
import random
def test(d, n):
    a = 2 + random.randint(1, n - 4)
    x = pow(a, d, n)
    if x == 1 or x == n - 1:
        return True
    while(d != n - 1):
        x = (x * x) % n
        d *= 2
        if x == 1:
            return False
        if x == n - 1:
            return True
    return False

def is_prime(n, k):
    if n <= 1 or n == 4:
        return False
    if n <= 3:
        return True
    
    d = n - 1
    while(d % 2 == 0):
        d //= 2
    
    for i in range(k):
        if(test(d, n) == False):
            return False
    return True

def factor(n):
    factors = []
    i = 1
    while i * i <= n:
        if n % i == 0:
            factors.append(i)
            if i != n // i:
                factors.append(n // i)
        i += 1
    factors.sort()
    print(' '.join(map(str, factors)))

def sum_factor(n):
    summ = 0
    i = 1
    while i * i <= n:
        if n % i == 0:
            summ += i
            if i != n // i:
                summ += n // i
        i += 1
    return summ

#n = 9876543210
number = int(input())
k = 3
if is_prime(number, k):
    print("Prime")
else:
    print("Not Prime")
    factor(number)
    print(sum_factor(number))
```

```python
import math

def mean_and_SD(score):
    mean = sum(score) / len(score)
    variance = sum(((x - mean) ** 2) for x in score) / len(score)
    SD = math.sqrt(variance)
    return mean, SD

def bestt(students, scores):
    highest = 0
    best_student = ''

    for i, student in enumerate(students):
        mean, SD = mean_and_SD(scores[i])
        if mean > highest:
            highest = mean
            best_student = student
            
    print(f"\nStudent with the highest average score is {best_student} with an average score of {highest:.2f}")

students = ['John Smith', 'Emily Johnson', 'Michael Williams', 'Jessica Brown', 'David Jones', 'Sophia Davis', 'Daniel Martinez', 'Olivia Miller', 'James Wilson', 'Emma Taylor']
scores = [[85, 78, 92, 88, 76, 90, 87, 89], [92, 86, 78, 94, 89, 91, 85, 82], [75, 80, 88, 84, 79, 92, 86, 90], [89, 92, 75, 81, 87, 79, 84, 88], [78, 81, 89, 93, 85, 90, 88, 82], [92, 85, 78, 90, 83, 89, 87, 91], [86, 91, 75, 80, 88, 84, 82, 89], [78, 84, 89, 92, 86, 91, 83, 80], [90, 88, 82, 79, 91, 76, 83, 85], [87, 80, 91, 85, 78, 92, 89, 84]]

bestt(students, scores)
```
