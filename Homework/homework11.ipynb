```python
class Book():
    def __init__(self, title, author, published_year):
        self.title = title
        self.author = author
        self.published_year = published_year
    
    def get_info(self):
        print(f"Title: {self.title}\nAuthor: {self.author}\nPublished year: {self.published_year}")
        
class Library:
    def __init__(self, book):
        self.book = book 
    
    def display_book_info(self):
        self.book.get_info()

book1 = Book("Hihi", "HiHI", 2000)
library = Library(book1)
library.display_book_info()
```

```python
import random

class CustomRandom(random.Random):
    def __init__(self):
        super().__init__()

    def get_custom_random_int(self, min_val, max_val) -> int:
        return self.randint(min_val, max_val)
    
    def random(self) -> int: 
        return int(super().random() * 100)
        

exrand = CustomRandom()
print(exrand.get_custom_random_int(10, 20)) 
print(exrand.random())
```
