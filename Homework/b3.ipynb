### Ex 5

```python
# num must be less than 10^18, otherwise the number will be too large to store in an int
num = 1112312304120844

# Store the prime factors and their counts in a list of tuples
primes = []

# Start with 2
div = 2

# You can use some fancy algorithms to test is the number a prime number
# or filter out the non-prime divisors, but this is the simplest way.

# Worst case scenario is that the number is a big prime number -> O(sqrt(n))
# Average case scenario is that the number is a composite number ~ O(log(n))

# Loop until the square root of the number is less than the divisor
while math.sqrt(num) >= div:
    
    # If the number is divisible by the divisor, divide the number by the divisor and count
    # the number of times it is divisible
    if num % div == 0:
        count = 0
        while num % div == 0:
            
            num //= div
            count += 1
        primes.append((div, count))
        
    # If the number is not divisible by the divisor, increment the divisor
    # Divisors will always be a prime number
    div += 1

# The remaining number is a big prime number
if num != 1:
    primes.append((int(num), 1))
print(primes)
        
```

Recall the probability theory

```python
nums_factor = 1

# The number of factors of a number is the product of the count of the prime factors plus one
for prime, count in primes:
    nums_factor *= (count + 1)
print(nums_factor)
```

```python
tol_sums = 1

# The sum of the divisors of a number is the product of the sum of the powers of the prime factors
for prime, count in primes:
    sums = 0
    for i in range(count + 1):
        sums += prime ** i
    tol_sums *= sums
print(tol_sums)
```

Test to reconstruct the original number

```python
original = 1
for prime, count in primes:
    original *= prime ** count
print(original)
assert original != num, "The original number is not the same as the number inputted"
```

### Ex7* Longest Increasing Subsequence (LIS)

O(n^2) solution

```python
a = [10,9,2,5,3,7,101,1,18]
dp = [0] * len(a)
trace = [-1] * len(a)

dp[0] = 1
for i in range(1, len(a)):
    for j in range(0, i):
        if dp[i] < dp[j] and a[i] > a[j]:
            dp[i] = dp[j]
            trace[i] = j
    dp[i] += 1
    
index_max = 0
for i in range(1, len(a)):
    if dp[i] > dp[index_max]:
        index_max = i

sub = []
i = index_max
while i != -1:
    sub.append(a[i])
    i = trace[i]

print(dp[index_max])
print(sub[::-1])
```

O(nlog(n)) solution

```python
nums = [10, 9, 2, 5, 3, 7, 101, 1, 18] # The input list Output: [2, 3, 7, 101]
trace = [-1] * len(nums) # Store the previous index of the number
lis = [(nums[0],0)] # Store the lis of each length (not the final answer) (number, index)

# Misconception: The lis array is not the final answer, it is just to store the smallest number of each length

# lis : [1 , 3, 7, 18]  (min value of lis of length i)
#        1   2  3   4   (length of lis)  

# Concept solution:
# 1. For each number, find the smallest number in lis that is greater than the number
# 2. If the number is greater than the last number in the lis, append the number to the lis
# 3. If the number is not greater than the last number in the lis, find the smallest number 
#       that is greater than the number (using binary search)

# 4. Replace the number with the smallest number that is greater than the number (our aim is to 
#       find the smallest lis in each position, so other numbers can construct a longer lis)
    
#       e.g Given a list [1,5,3,4] , when selecting 3, our lis array are [1,5]. We will replace 
#       5 with 3, which means the increasing subsequence with length = 2 is [1,3] instead of [1,5],
#       so that the number 4 can later be added to the lis

# 5. Repeat until the end of the list 

# For modifying the trace array, we are no longer using the index 'j' like in O(n^2) solution, so we
# have to store it in a different way. 

# Since the trace pointed to the 2nd last element of increasing subsequence, we can store the index in 
# lis array, and the trace array will get the value `j` from the lis array

for i in range(1,len(nums)):
    
    # If the number is greater than the last number in the lis, append the number to the lis
    if nums[i] > lis[-1][0]:
        
        # Trace of it will be the last index of the lis
        trace[i] = lis[-1][1]
        lis.append((nums[i],i))   
    else:
        
        # Insert the number in the lis array
        l, r = 0, len(lis)-1
        while l < r:
            mid = (l+r)//2
            if lis[mid][0] < nums[i]:
                l = mid+1
            else:
                r = mid
                
        # If the inserted number is the smallest number in the lis, the trace will be -1
        if l == 0:
            trace[i] = -1
        else:
            # The trace will be the index of the previous number in the lis
            trace[i] = lis[l-1][1]  
        
        # Replace the number with the smallest number that is greater than the number
        lis[l] = (nums[i],i)
arr = []

# Trace back the lis from the last index
p = lis[-1][1]
while p >= 0:
    arr.append(nums[p])
    p = trace[p]
    
print(arr[::-1])
print(lis)
```
